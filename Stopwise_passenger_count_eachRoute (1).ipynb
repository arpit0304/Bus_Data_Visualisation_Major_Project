```python
import pandas as pd
df=pd.read_csv('/content/bus_data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
df=df[['fromstopname', 'route_name', 'passengers']]
```

```python
df.shape
```

```python
# grouping the data by route name and bus-stop_name and sum of the number of passengers
result = df.groupby(['fromstopname','route_name'])['passengers'].sum()

print(result)
```

```python
result.to_csv("stopWise_data.csv")
```
