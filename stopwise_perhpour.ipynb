```python
value_counts_series = df['fromstopname'].value_counts()
sorted_value_counts = value_counts_series.sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(10, 6))  # Set the size of the figure

# Plot the bar graph
sorted_value_counts.plot(kind='bar')

plt.title('Value Counts of fromstopname')
plt.xlabel('fromstopname')
plt.ylabel('Count')

plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()       # Adjust layout for better spacing

plt.show()  # Display the plot
```

```python
for a in range(14, 15):


      b=a+1
      fr = '11/14/2022 '+str(a)+':00'
      to ='11/14/2022 '+str(b)+':00'
      df2 = df[(df['ticket_time'] >= fr) & (df['ticket_time'] <= to)].copy()

      value_counts_series = df2.groupby('fromstopname')['passengers'].sum()
      sorted_series = value_counts_series.sort_values(ascending=False)
      sorted_value_counts =  sorted_series[:40]

      plt.figure(figsize=(15, 9))  # Set the size of the figure

      sorted_value_counts.plot(kind='bar')

      plt.title(fr+"-"+to)
      plt.xlabel('fromstopname')
      plt.ylabel('passenger Count')

      plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
      plt.tight_layout()       # Adjust layout for better spacing

      plt.show()  # Display the plot
      plt.show()

```
